```python
import pandas as pd
```

```python
data = pd.read_csv("supply_chain_data.csv")
data.head()
```

```python
missing_values = data.isnull().any(axis=1)
print("Rows with missing values:")
print(missing_values)
```

```python
duplicate_values = data[data.duplicated()]
print("Duplicate Rows:")
print(duplicate_values)
```

```python
data.dropna(axis=0,inplace=True) # if you want to remove missing values (if there's any)
```

```python
data.drop_duplicates(inplace=True) # if you want to remove duplicates (if there's any)
```
